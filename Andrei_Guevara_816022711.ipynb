<h2>1.0 Data Cleaning - Attribute Downsampling:</h2> 

<p>
I have decided to overwrite the entire release date attribute with the year from the end of the title, and since this data is no longer needed at the end of the title, it will be removed. 

The missing values in the license column will be filled with the most frequent value of that category.

Since words will cannot be interpreted in graphs I have decided to remove the movie info column as well
</p>

```python
# The titles will be extracted, and appended to the end of the titles list

titles = []

for name in df["Title"].values.tolist():
    
    stripped = name[0:len(name)-7]
    titles.append(stripped)
    
# Now the titles become a numpy array, then we overwrite the titles with the new stripped titles    
titles = np.array(titles)
df['Title'] = titles
```

```python
most_frequent = df["License"].value_counts().idxmax() # Finding out the most frequent value in the column
df["License"].fillna(most_frequent, inplace= True) # Replacing the missing values with this value
```

```python
df.head()
```

```python
df = df.drop(columns="Movie Info")
```

Conversion of Movie Runtime into minutes

```python
times = df['Movie Runtime'].values.tolist()

minutes = []

for time in times:
    
    hour = int(time.split(" ")[0]) * 60
    minute = 0
    
    if len(time.split(" ")) > 2:
        
        minute = int(time.split(" ")[2])
    
    minutes.append(hour+minute)
    
minutes = np.array(minutes)
df['Movie Runtime'] = minutes

df.head()
```

```python
df1 = df.copy(deep=True)
```

<h2>1.5 Data Cleaning - One hot encoding the genre feature</h2>
<p>One hot encoding the Genre column</p>

```python
# Turning the Genre column into a list
genre = df['Genre'].str.split(',').tolist()

# Turning the list of lists of genres into one list
flat_genre = [item.replace("[","").replace("]","").replace("'","").strip() for sublist in genre for item in sublist]

# Removing duplications
set_genre = set(flat_genre)

# Turning the set into a list
unique_genre = list(set_genre)

# Storing 1 or 0 if the movie is in the genre
for genre in unique_genre:
    df[genre] = df['Genre'].str.contains(genre).astype('int')

# Deleting the original Genre column    
df.drop('Genre', axis=1, inplace=True)

df
```
